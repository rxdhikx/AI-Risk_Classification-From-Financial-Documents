# <b> Results from 2023 Financial Reports (Q1, Q2, Q3, Q4)</b>

```python
#testing QTR1 first

import pandas as pd
df = pd.read_csv("/Users/rravindra0463/Desktop/ai_proj/venv/output/Results_QTR1.csv")
df.describe()
```

So about 734 files are ones which have AI related keywords in them.

```python
# check how many files orginally present in the concerned folder,i.e QTR1 here
import os
folder_path = '/Users/rravindra0463/Desktop/ai_proj/venv/QTR1'

# list all files in the folder
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# count the number of files
num_files = len(files)
print(f'Number of files in the folder QTR1: {num_files}')

# calculate metrics
unique_filenames = df['filename'].nunique()
percentage = (unique_filenames/ num_files) * 100

print(f'Number of relevant files extracted: {unique_filenames}')
print(f'Percentage of relevant files: {percentage:.2f}%')
```

So out of 7k+ financial reports, about 730 are those whose risk factors are related to AI.

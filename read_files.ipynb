Let's see what unique keywords have been extracted to double-verify if they are AI related:

```python
# Initialize an empty set to store unique keywords
unique_keywords = set()

# Iterate through each row in the 'KeywordMentioned' column
for keywords in df['KeywordMentioned'].dropna():
    # Split the keywords by comma and strip any extra whitespace
    split_keywords = [keyword.strip() for keyword in keywords.split(',')]
    # Add the split keywords to the set
    unique_keywords.update(split_keywords)

# Print each unique keyword
for keyword in unique_keywords:
    print(keyword)
```
